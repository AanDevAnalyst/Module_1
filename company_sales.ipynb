```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_excel("Company_Sales_Data.xlsx")
```

```python
df.head(2)
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(12,8))
plt.plot(df["Month_Number"], df["Total_Profit"], marker='o', linestyle='-', color='b')

plt.title("Company Sales Profit Per Month", fontsize=14)
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.xticks(df["Month_Number"])
plt.yticks([]) 
plt.grid(True, linestyle="--", alpha=0.6)

for x, y in zip(df["Month_Number"], df["Total_Profit"]):
    plt.text(x, y, str(y), ha="center", va="bottom", fontsize=10)

plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.subplot(1, 2, 1)
plt.plot(df["Month_Number"], df["Bathingsoap"], marker='o', color="blue")
plt.title("Monthly Sales of Bathingsoap")
plt.xlabel("Month Number")
plt.ylabel("Units Sold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.yticks([])  

for x, y in zip(df["Month_Number"], df["Bathingsoap"]):
    plt.text(x, y, str(y), ha="center", va="bottom", fontsize=10)

plt.subplot(1, 2, 2)
plt.plot(df["Month_Number"], df["Face_Wash"], marker='s', color="green")
plt.title("Monthly Sales of Face Wash")
plt.xlabel("Month Number")
plt.ylabel("Units Sold")
plt.grid(True, linestyle="--", alpha=0.6)
plt.yticks([]) 

for x, y in zip(df["Month_Number"], df["Face_Wash"]):
    plt.text(x, y, str(y), ha="center", va="bottom", fontsize=10)

plt.tight_layout()
plt.show()
```
